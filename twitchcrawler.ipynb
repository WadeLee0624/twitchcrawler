```python
from bs4 import BeautifulSoup
import requests
import re  
import csv
import time
import threading
from pandas import DataFrame
from pandas import read_csv
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains #引入鼠标
from selenium.webdriver.common.keys import Keys #引入键盘


headers = {
    'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1"
}  


def top():                      #此函式找出目前觀看人數前幾名LOL實況，回傳開台者名單
    onstreaming = []
    url="https://www.twitch.tv/directory/game/League%20of%20Legends"  
    driver.get(url)
    time.sleep(3)
    search_filter = driver.find_element_by_xpath('//input[@placeholder="搜尋標籤"]')   #搜尋英文標籤，目的是只抓英文頻道
    search_filter.send_keys(u"英文")
    time.sleep(1)
    Englinshlabel = driver.find_element_by_xpath('//button[@tabindex="-1"]')
    Englinshlabel.click()
    time.sleep(1)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    top_soup = BeautifulSoup(driver.page_source,'lxml')  #第一次寫入soup
    streamer = top_soup.find_all('a',attrs={"data-a-target": "preview-card-title-link"})
    for i in range(2):
        onstreaming.append("https://www.twitch.tv"+streamer[i].attrs['href'])
    driver.close()
    return onstreaming


def crawler(stream_url):
    threads = []  
    print(stream_url)
    for i in range(len(stream_url)):
        threads.append(threading.Thread(target = crawler_tread, args = (stream_url[i],)))    
        threads[i].start()
        time.sleep(5)
    
    
def crawler_tread(url):
    driver1 = webdriver.Chrome('/Users/lwcoo/Downloads/chromedriver')
    driver1.get(url)
    time.sleep(3)
    soup = BeautifulSoup(driver1.page_source,'lxml')  #第一次寫入soup
    viewer_cnt = soup.find('div',class_='tw-mg-l-05 tw-stat__value')  #先抓取人數
    print("觀眾數:"+viewer_cnt.string)
    print("觀眾ID:")
    driver1.find_element_by_xpath('//button[@aria-label="Users in Chat"]').click() #點擊觀眾按鈕
    time.sleep(3)
    roll_list = driver1.find_element_by_xpath("//div[@class='simplebar-scroll-content chat-viewers__scroll-container']") #找到觀眾list卷軸
    roll_list.click()  #先讓鼠標點擊一下
    x = int(viewer_cnt.string.replace(',', ''))/25   #特殊字元替換後用人數/25看要跑幾圈 
    for i in range(1,int(x)):  #此迴圈滾動觀眾在線清單畫面
        ActionChains(driver1).key_down(Keys.PAGE_DOWN).perform()  
        time.sleep(0.1)
    soup = BeautifulSoup(driver1.page_source,'lxml')    #在寫入一次soup覆蓋，因為滾動過
    soup1 = soup.find('div',attrs={"aria-labelledby": "chat-viewers-list-header-使用者"})  #soup1找到使用者清單位置
    viewer = soup1.find_all('button',class_='chat-viewers-list__button')   #所有按鈕中的使用者id寫入viewer
    for x in viewer:   
        print(x.string)   #印出使用者ID


driver = webdriver.Chrome('/Users/lwcoo/Downloads/chromedriver')
driver.set_window_size(1000,5000)
onstreaming = top()
crawler(onstreaming)



```
